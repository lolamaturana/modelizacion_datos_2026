```python
# PREPROCESAMIENTO DE LOS DATOS.
import pandas as pd
df_train = pd.read_csv("data/df_train_small.csv")
```

```python
# workflow de preprocesamiento.
# 1.- Procesar nulls/nan
# 2.- Procesar variables categoricas
     # 2.1 baja cardinalidad
     # 2.2 alta cardinalidad - generalmente serán texto.
# 3.- Procesar variables numericas: normalizacion, transformacion, outliers.
# 4.- Transformación de variables: crear nuevas variables a partir de las existentes. (feature engineering)
# 5.- Selección de variables: eliminar variables que no aportan información relevante para el modelo. (feature selection)

```

```python
# Trabajamos con los nulls/nan
# dataframe de nulls de cada variable y su porcentaje respecto al total de filas ordenados de mayro a menor .eliminamos las que no tengan null

nulls = pd.DataFrame(df_train.isna().sum(), columns=["nulls"])
nulls["percent"] = nulls["nulls"] / len(df_train) * 100
nulls = nulls.sort_values("nulls", ascending=False)
nulls.query("nulls > 0").shape
```
